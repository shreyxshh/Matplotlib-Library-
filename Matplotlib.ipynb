```python
# importing matplotlib

import matplotlib.pyplot as plt
```

```python
# Line plot

x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x,y)

plt.xlabel("This is x-label")
plt.ylabel("This is y-label")
plt.title("This is our first Line Plot")

plt.show()
```

```python
plt.figure(figsize=(6,4))   # h = 6px, b = 4px

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.plot(x,y1, label = 'Line 1', color="violet")
plt.plot(x,y2, label = 'Line 2', color="red")

plt.legend()
plt.grid()
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.subplot(2,2,1)      # .subplot(rows,cols,quadrant)
plt.plot(x,y1,'y--')    # y-- -> representaion in dashesd yellow line(any color)

plt.subplot(2,2,4)
plt.plot(x,y2)
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
fig = plt.figure(figsize=(6,4))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])     # [left,bottom,width,height]
axes1.plot(x,y)
plt.title("Axes1")

axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes2.plot(y,x)
plt.title("Axes2")

# axes3 = fig.add_axes([0.5,0.2,0.3,0.3])
# axes2.plot(y,x)
# plt.title("Axes3")
```

```python
# DPI -> dots per inch => controls zoom in or zoom out for a graph

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.figure(figsize=(6,4), dpi=100)
plt.grid()
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,4))

x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.subplot(2,2,1)
plt.plot(x,y1,label = 'blue' , color = 'blue')
plt.title('Axes1')

plt.subplot(2,2,4)
plt.plot(x,y2,label = 'yellow', color = 'yellow' )
plt.title('Axes2')
```

```python
fig = plt.figure(figsize=(6,4))

x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.plot(x,y1,label = 'blue' , color = 'blue', lw = '2.5')
plt.plot(x,y2,label = 'green', color = 'green', lw = '1.5', ls='-.' , marker = '*')


plt.title('Graph')
plt.legend()
```

```python
# Scatter ,Histogram & Boxplot

x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]
# 1. Scatter plot -> 2 columns/variables/features

plt.scatter(x,y)
```

```python
# 2. Histogram plot -> shows spread of only on column/variable

import numpy as np

nums = np.random.randn(100)
plt.hist(nums, bins=20)
```

```python
from typing_extensions import ValuesView
# 3. Box plot -> shows outliers of single column

plt.boxplot(nums)

# min
# 25% -> Lower Quartile
# 50% -> Median
# 75% -> Upper Quartile
# max

# Values between lower and uppre quartile are generic and actual Values
# Values outside lower and uppre quartile are considered to be outliers
```
